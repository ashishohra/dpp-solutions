```python
# 11
import pandas as pd
import numpy as np
data = {'student': ['ravi','kavi','shavi'], 'score':[88,np.nan,98]}
df1 = pd.DataFrame(data)
print(df1)
dfdropped = df1.dropna(subset=["score"])
print(dfdropped)
df2 = pd.DataFrame(data)
scoremean =  df2["score"].mean()
df2["score"] = df2["score"].fillna(scoremean)
print(scoremean)
print(df2)

```

```python
# 12
import pandas as pd
data = {
    'Store': ['North', 'South', 'North', 'West', 'South', 'North'],
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Laptop', 'Mouse'],
    'Sales': [1200, 25, 300, 75, 1150, 30]
}
df = pd.DataFrame(data)
print(df)
total_sales = df.groupby("Store")["Sales"].sum()
print(total_sales)
```

```python
#13
import pandas as pd
employees_data = {
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
}
df_employees = pd.DataFrame(employees_data)
print(df_employees)
salaries_data = {
    'Emp_ID': [101, 102, 103, 105],  # 104 is missing here, 105 is new
    'Salary': [60000, 45000, 70000, 55000]
}
df_salaries = pd.DataFrame(salaries_data)
print(df_salaries)
merged_df = pd.merge(df_employees, df_salaries, on="Emp_ID", how="inner")
print(merged_df)
```

```python
#14
import pandas as pd
import numpy as np
data = {
    'Student': ['Grace', 'Heidi', 'Ivan', 'Judy', 'Kevin'],
    'Final_Score': [78, 45, 50, 92, 33]
}
df = pd.DataFrame(data)
print(df)
df["Status"] = np.where(df["Final_Score"] >= 50, "Pass", "Fail")
print(df)
```

```python
#2
from scipy import constants
print("Pi:", constants.pi)
print("Speed of light:", constants.c)
print("Planck's constant:", constants.h)
```

```python
# 3
from scipy import constants
distance_meters = 55 * constants.mile
print("55 miles in meters:", distance_meters)
weight_kilograms = 10 * constants.pound
print("10 pounds in kilograms:", weight_kilograms)
```

```python
#4
from scipy import special
import numpy as np
arr = np.array([1, 8, 27, 64])
cubic_roots = special.cbrt(arr)
print("Cubic roots:", cubic_roots)
```

```python
#5
from scipy import stats
import numpy as np
dataset = np.array([1, 2, 2, 3, 4, 4, 4, 5, 6])
mode_result = stats.mode(dataset)
print("Mode:", mode_result.mode)
print("Count:", mode_result.count)
```

```python
#6
from scipy import stats
numerical_array = [2, 8, 32]
g_mean = stats.gmean(numerical_array)
print("Geometric Mean:", g_mean)
```

```python
#7
from scipy import linalg
import numpy as np
matrix = np.array([[4, 5], [2, 3]])
det = linalg.det(matrix)
print("Determinant:", det)
```

```python
#8
from scipy import linalg
import numpy as np
M = np.array([[1, 2], [3, 4]])
M_inverse = linalg.inv(M)
print("Inverse Matrix:\n", M_inverse)
```

```python
#9
from scipy import stats
import numpy as np
scores = np.array([55, 60, 65, 70, 75, 80, 85, 90, 95, 100])
percentile_75 = stats.scoreatpercentile(scores, 75)
print("75th Percentile:", percentile_75)
```

```python
#10
from scipy import stats
import numpy as np
measurements = np.array([10, 20, 30, 40, 50])
z_scores = stats.zscore(measurements)
print("Z-scores:", z_scores)
```
